# Quiz 7 เปรียบเทียบตำรวจที่เป็น White และ Black ชอบเรียกค้นคนอายุเท่าไหร่ และเชื้อชาติไหน ต่างกันอย่างไร

````python
# Filter for White officers
white_officer_stops = SS_warwick[SS_warwick['Officer-defined ethnicity'] == 'White']

# Group by Age range and Self-defined ethnicity for White officers
white_officer_analysis = white_officer_stops.groupby(['Age range', 'Self-defined ethnicity']).size().reset_index(name='count')
print("Stops by White Officers:")
display(white_officer_analysis)

# Filter for Black officers
black_officer_stops = SS_warwick[SS_warwick['Officer-defined ethnicity'] == 'Black']

# Group by Age range and Self-defined ethnicity for Black officers
black_officer_analysis = black_officer_stops.groupby(['Age range', 'Self-defined ethnicity']).size().reset_index(name='count')
print("\nStops by Black Officers:")
display(black_officer_analysis)
````

ตำรวจผิวขาวมีแนวโน้มที่จะเรียกตรวจค้นบุคคลในกลุ่ม White - English/Welsh/Scottish/Northern Irish/... ในทุกช่วงอายุ แต่มีการเรียกตรวจค้นกลุ่มอายุ 10-17 ปี และ over 34 ปี มากที่สุด

ตำรวจผิวดำมีจำนวนการเรียกตรวจค้นที่ค่อนข้างน้อยเมื่อเทียบกับตำรวจผิวขาว และมีการเรียกตรวจค้นหลากหลายเชื้อชาติ แต่มีแนวโน้มที่จะเรียกตรวจค้นกลุ่มคนผิวสีในบางช่วงอายุ

###1.1 Drop เฉพาะ column  ที่มีค่า missing มากกว่า 50%


````python
# Calculate the percentage of missing values for each column
missing_percentage = SS_warwick.isnull().sum() / len(SS_warwick) * 100

# Identify columns with more than 50% missing values
columns_to_drop = missing_percentage[missing_percentage > 50].index

# Drop the identified columns
SS_warwick_dropped = SS_warwick.drop(columns=columns_to_drop)

# Display the first few rows of the modified DataFrame
display(SS_warwick_dropped.head())
````

````python
SS_warwick_dropped.dropna()
````

### 1.2 Drop เฉพาะ Data ที่มี Missing ใน Colunm ที่เราสนใจ

````python
SS_warwick.dropna(subset=['Date','Gender','Age range','Self-defined ethnicity','Officer-defined ethnicity','Object of search'])
````

## 2 เติมค่า missing `fillna()`

## 2.1 เติมค่าด้วย dummy value (ซึ่งต้องเป็นตัวเลขที่ไม่มีอยู่ในข้อมูล เพื่อสื่อว่าเป็น ค่า missing)

````python
set(SS_warwick['Part of a policing operation'])
````

````python
SS_warwick.fillna({'Part of a policing operation':False})
````

````python
SS_warwick.fillna({'Part of a policing operation':False}, inplace=True)
````

## 2.2เติม missing ด้วยค่ากลาง

เติม Latitude และ Longtitude ด้วยค่าเฉลี่ย (Mean)

````python
SS_warwick.fillna({'Part of a policing operation':False,
                   'Latitude':SS_warwick['Latitude'].mean(),
                   'Longitude':SS_warwick['Longitude'].mean()
                   })
````
